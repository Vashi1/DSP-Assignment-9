```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

#p1 = 5
print("\n\t\t\tImplementing PCA with 5 Components\t\t\t\n")
pca = PCA(n_components = 5, random_state = 42)
scaler = StandardScaler()
pipeline = Pipeline([('scaler', scaler), ('pca', pca)])
pipeline.fit(train)
print("\n\t\t\tExplained Variance Ratio for Scaled Train Set\n\t\t\t")
print("\n\t\t", pca.explained_variance_ratio_, "\t\t\n")
print("\n\t\t\tExplained Singluar Values\t\t\t\n")
print("\n\t\t", pca.singular_values_, '\t\t\n')

#p2 = 8
print("\n\t\t\tImplementing PCA with 8 Components\t\t\t\n")
pca1 = PCA(n_components = 8, random_state = 42)
scaler = StandardScaler()
pipeline = Pipeline([('scaler', scaler), ('pca', pca1)])
pipeline.fit(train)
print("\n\t\t\tExplained Variance Ratio for Scaled Train Set\n\t\t\t")
print("\n\t\t", pca1.explained_variance_ratio_, "\t\t\n")
print("\n\t\t\tExplained Singluar Values\t\t\t\n")
print("\n\t\t", pca1.singular_values_, '\t\t\n')
```

*Subpart 7*

Implementing K-Means Clustering Algorithms using Elbow Method


```python
#Implementing K-Means Clustering Algorithm
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from scipy.spatial.distance import cdist

fits = []
score = []
distortions = []
intertias = []

K = range(1, 20)

for k in K:
    #Building the K-Means Model
    kmm = KMeans(n_clusters = k, random_state = 42, n_init = 20).fit(train)
    kmm.fit(train)
    distortions.append(sum(np.min(cdist(train, kmm.cluster_centers_, 'euclidean'), axis = 1)) / train.shape[0])
    intertias.append(kmm.inertia_)
```

Plotting the Graphs indicating the optimal number of clusters using Elbow Method and Intertia

```python
plt.plot(K, distortions, 'bx-')
plt.xlabel("Values of K")
plt.ylabel("Distortion")
plt.title("Elbow Method using Distrotion")
xr = list(range(1, 20))
plt.xticks(xr, K)
plt.show()

plt.plot(K, intertias, 'bx-')
plt.xlabel("Values of K")
plt.ylabel("Interia")
plt.xticks(xr, K)
plt.title("Elbow Method using Intertia")
plt.show()
```

```python
km = KMeans(n_clusters = 3, random_state = 42, n_init = 20).fit(train)
plt.scatter(train.iloc[:, 0], train.iloc[ :, 1])
plt.scatter(km.cluster_centers_[:,0], km.cluster_centers_[:, 1], c = 'red', marker = 'x')
plt.show()
```

Implementing K-Means Clustering Algorithm using Silhoute Score

```python
silhouette_coefficients = []
for k in range(2, 21):
    km_s = KMeans(n_clusters = k, random_state = 42, n_init = 20)
    km_s.fit(train)
    score = silhouette_score(train, km_s.labels_)
    silhouette_coefficients.append(score)
plt.plot(range(2, 21), silhouette_coefficients)
plt.title("Plot of Silhouette Scores")
plt.xlabel("Number of Clusters")
xval = list(range(2, 21))
plt.xticks(xval, range(2, 21))
plt.ylabel("Silhouette Score")
plt.show()
```
